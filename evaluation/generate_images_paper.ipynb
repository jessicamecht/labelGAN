```python
import os 
from PIL import Image, ImageOps
import matplotlib.pyplot as plt 
import numpy as np
```

```python
real_image_path = "/data3/jessica/data/labelGAN/train_images/imgs/"
generated_image_path = "/data3/jessica/data/labelGAN/train_images/generated/"
masks_propagated = "/data3/jessica/data/labelGAN/train_images/masks/"
```

```python
imgs = os.listdir(real_image_path)[9:19]
fig, axs = plt.subplots(3, 10, figsize=(19.6, 6), gridspec_kw={'wspace': 0, 'hspace': 0})
plt.subplots_adjust(wspace=0, hspace=0)
for i, elem in enumerate(imgs):
    mask_name = elem.replace(".png", "_mask.jpg")
    im_p = real_image_path + elem
    img = Image.open(im_p)
    img = img.resize((1024, 1024))
    img = np.array(ImageOps.grayscale(img))
    
    mask_p = masks_propagated + mask_name
    mask = Image.open(mask_p)
    mask = np.array(ImageOps.grayscale(mask))
    img_generated = Image.open(generated_image_path + elem)
    img_generated = np.array(ImageOps.grayscale(img_generated))

    axs[0, i].imshow(img, cmap='gray', aspect='equal')
    axs[0, i].axis('off')

    axs[1, i].imshow(img_generated, cmap='gray', aspect='equal')
    axs[1, i].axis('off')

    axs[2, i].imshow(mask, cmap='gray', aspect='equal')
    axs[2, i].axis('off')
fig.savefig("generations_and_embeddings.pdf")
```
