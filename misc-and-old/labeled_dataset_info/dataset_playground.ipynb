```python
with_label_count = 0
labeled_images = []
for image_path in image_paths:
    for label_path in label_paths:
        if image_path == label_path.replace('_mask', ''):# and "_mask" in label_path:
            labeled_images.append(image_path)
            with_label_count +=1
```

```python
with_mask_in_labelname_count = 0
for image_path in image_paths:
    for label_path in label_paths:
        if image_path == label_path.replace('_mask', '') and "_mask" in label_path:
            print(image_path)
            with_mask_in_labelname_count +=1
```
